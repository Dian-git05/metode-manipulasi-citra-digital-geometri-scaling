```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def scale_image(image, scale_x, scale_y, interpolation=cv2.INTER_LINEAR):
    """
    Melakukan penskalaan citra berdasarkan faktor skala X dan Y.

    Parameters:
    - image: array citra (numpy ndarray)
    - scale_x: faktor skala horizontal
    - scale_y: faktor skala vertikal
    - interpolation: metode interpolasi (default: INTER_LINEAR)

    Returns:
    - scaled_image: citra hasil penskalaan
    """
    if image is None:
        raise ValueError("Citra tidak valid atau belum dimuat.")

    height, width = image.shape[:2]
    new_width = int(width * scale_x)
    new_height = int(height * scale_y)

    scaled_image = cv2.resize(image, (new_width, new_height), interpolation=interpolation)
    return scaled_image
```

```python
# Ganti dengan path gambar lokal kamu
image_path = "WhatsApp Image 2025-11-12 at 11.15.52.jpeg"
image = cv2.imread(image_path)

# Konversi BGR ke RGB untuk ditampilkan dengan matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Gambar Asli")
plt.axis("off")
plt.show()
```

```python
# Scaling dengan faktor 1.5 horizontal dan 0.5 vertikal
scaled = scale_image(image, scale_x=1.5, scale_y=0.5)

# Konversi BGR ke RGB
scaled_rgb = cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB)

plt.imshow(scaled_rgb)
plt.title("Hasil Scaling (1.5x, 0.5y)")
plt.axis("off")
plt.show()
```
